```python
#create a list of Boroughs that we want to delete in next step, this list will be used as condition to subset
nonBoroughs = ['Inner London', 'Outer London', 
               'NORTH EAST', 'NORTH WEST', 'YORKS & THE HUMBER', 
               'EAST MIDLANDS', 'WEST MIDLANDS',
              'EAST OF ENGLAND', 'LONDON', 'SOUTH EAST', 
              'SOUTH WEST', 'England']
```

```python
#Create a new column year with a lambda function that gets the year from the date on Month column
#Ask Dr. Maskall about this error
df['Year'] = df['Month'].apply(lambda t: t.year)
```

```python
#using groupby method on df to aggregate by Borough and Year and get the mean from them
dfg = df.groupby(by=['London_Borough', 'Year']).mean()
#get a sample of 10 observations
dfg.sample(10)
```

```python
#rest index of df because we can't make calculations with indexes
dfg = dfg.reset_index()
#check if it has worked
dfg.head()
```

```python
#3. Modeling
#create a function that divides one variable between another
#first variable is year 1998 avg prices - we calculate it as average price from d with condition year 1998
#first variable is year 2018 avg prices - we calculate it as average price from d with condition year 2018
def create_price_ratio(d):
    y1998 = float(d['Average_price'][d['Year']==1998])
    y2018 = float(d['Average_price'][d['Year']==2018])
    ratio = [y2018/y1998]
    return ratio
```

```python
#test of the function with one Borough to see if it works
create_price_ratio(dfg[dfg['London_Borough']=='Barking & Dagenham'])
```

```python
#to apply this function to all Boroughs we will need to store results of function somewhere, we create an empty dictionary for this purpose.
final = {}
```

```python
#we make a loop that goes on unique Boroughs and makes the following
for b in dfg['London_Borough'].unique():
    #create a variable that stores the value of the key
    borough = dfg[dfg['London_Borough'] == b]
    #makes an entry in hte ditionary  with this key, key = unique Borough, value = function applied to that key
    final[b] = create_price_ratio(borough)
# we check if it worked by printing the dictionary, that now is populated with data
print(final) 
```

```python
#we make a new variable to store the dictionary, but in data frame type
df_ratios = pd.DataFrame(final)
```

```python
#preview the new variable
df_ratios.head()
```

```python
#as before, we need to transpose and reindex the dataframe to be able to use the Boroughs
df_ratios_T = df_ratios.T
df_ratios = df_ratios_T.reset_index()
#check if the transformations worked
df_ratios.head()
```

```python
# Let's just rename the 'index' column as 'London_Borough', and the '0' column to '2018'.
df_ratios.rename(columns={'index':'Borough', 0:'2018'}, inplace=True)
df_ratios.head()
```
